### Data pre-processing

```python
import pandas as pd
import numpy as np
import os.path
import os
import re
```

```python
#importing files
parent_dir = '/home/Time Series/dataset'

xlsx_files = [os.path.join(parent_dir, xlsx) for xlsx in os.listdir(parent_dir) if os.path.isfile(os.path.join(parent_dir, xlsx)) and xlsx.endswith('.xlsx')] 

li = []
   
for filename in xlsx_files:
    df = pd.read_excel(filename,index_col=None,header=[1],usecols=["Name","Age","Sex","Area","Date Of Reporting"],
                        parse_dates=['Date Of Reporting'])
    df['Date Of Reporting'] = pd.to_datetime(df['Date Of Reporting'],errors = 'coerce')
    li.append(df)  

data = pd.concat(li)

#drop rows with no names & date of reporting
data = data[data['Name'].notna()]
data = data[data['Date Of Reporting'].notna()]
data = data.reset_index(drop=True)
```

```python
#remove html-tags, and non-alphabetic characters 
def keepAlpha(sentence):
    alpha_sent = ""
    for word in sentence.split():
        alpha_word = re.sub('[^a-z A-Z]+', ' ', word)
        alpha_sent += alpha_word
        alpha_sent += " "
    alpha_sent = alpha_sent.strip()
    return alpha_sent

#remove any punctuation or special characters
def cleanPunc(sentence): 
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    cleaned = cleaned.strip()
    cleaned = cleaned.replace("\n"," ")
    return cleaned

#remove white space  
def whitespace(sentence):
    space_removed = " ".join(sentence.split())
    return space_removed
```
